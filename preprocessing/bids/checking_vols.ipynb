```python
# setting up and checking vols
import glob, os
import subprocess

qa_dict = {}

NIF_subjs = glob.glob('/projects/niblab/bids_projects/Experiments/Bevel/Nifti/sub-*')
for sub in sorted(NIF_subjs):
    #Niftis
    _id = sub.split("/")[-1]
    n = glob.glob(os.path.join(sub, "func/*.nii.gz"))
        #print(x)
        # test case 
    runs = [i for i in n if "task-prob" in i.split("/")[-1].split("_") ]
    if len(runs) != 4:
        print("FOUND BAD", _id)
    for y in runs:
        run =  y.split("/")[-1].split("_")[2]
        print("5 differrent runs found for SUBJECT ", y.split("/")[-1].split("_")[0])
        if not _id in qa_dict:
            qa_dict[sub.split("/")[-1]] = {
                    "VOL_COUNT" : [],
                    "BAD_RUNS": []
            }
               # if "rest" in n.split("/")[-1].split("_"):
                #### VOLUME PARAMETER 
        vol = int(subprocess.check_output(['fslnvols', y]))
        vol_count = int(subprocess.check_output(['fslnvols', y]))
        #if vol_count != 199:
            #print("Error in volume RUN....",run)
        print("FILE ", y)
        print("VOLUME: ", vol_count)
        qa_dict[_id]["VOL_COUNT"].append([run, vol_count])
        if vol_count != 199:
            qa_dict[_id]["BAD_RUNS"].append(run)
                
for _id in qa_dict:
    # Check if list is empty 
    if not qa_dict[_id]["BAD_RUNS"]:
        pass
    elif len(qa_dict[_id]["BAD_RUNS"]) > 1:
        print("MULTIPLE: ", qa_dict[_id]["BAD_RUNS"])
        # if multiple run errors, see if there is enough to get all needed runs
        print("RUN COUNT: ",len(qa_dict[_id]["VOL_COUNT"]))
        for x in qa_dict[_id]["VOL_COUNT"]:
            print("%s"%(x))
    else:
        print("SINGLE: ", qa_dict[_id]["BAD_RUNS"])
        print("RUN COUNT: ",len(qa_dict[_id]["VOL_COUNT"]))
            
```
