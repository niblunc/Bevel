# Setting up the reduced ROI atlas

```python
import glob, os
import pandas as pd
```

```python
atlas_txt ='reduced_ROI_lst.csv'
```

```python
df = pd.read_csv('reduced_ROI_lst.csv', sep=',')
```

```python
df.head()
```

```python
df.drop(["MNI_x", "MNI_y", "MNI_z", "orig_order_ROI", "sep_ROI_num"], axis='columns', inplace=True)
df.head()
```

```python
df.rename({"Region": "regions", "beta_roi_index": "index"}, axis='columns', inplace=True)
df.head()
```

```python
atlas_tsv = atlas_txt.replace(".csv", ".tsv")
df.to_csv(atlas_tsv, sep="\t", index=False)
```

```python
atlas_df = pd.read_csv(atlas_tsv, sep="\t")
atlas_df
```
